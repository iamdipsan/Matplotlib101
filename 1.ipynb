```python
#colors(hex) and line(width and style)  and marker(size)
#labels title
plt.plot(batsman['index'],batsman['V Kohli'],color= 'green')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black')
plt.title('Rohit sharma vs virat kholi carrer comparison')
plt.xlabel('Season')
plt.ylabel('Runs scored')
```

```python

plt.plot(batsman['index'],batsman['V Kohli'],linestyle='dashed')
plt.plot(batsman['index'],batsman['RG Sharma'],linestyle='dashed')
plt.title('Rohit sharma vs virat kholi carrer comparison')
plt.xlabel('Season')
plt.ylabel('Runs scored')
```

```python

plt.plot(batsman['index'],batsman['V Kohli'],linestyle='dotted')
plt.plot(batsman['index'],batsman['RG Sharma'],linestyle='dotted')
plt.title('Rohit sharma vs virat kholi carrer comparison')
plt.xlabel('Season')
plt.ylabel('Runs scored')
```

```python

plt.plot(batsman['index'],batsman['V Kohli'],linestyle='dashdot')
plt.plot(batsman['index'],batsman['RG Sharma'],linestyle='dashdot')
plt.title('Rohit sharma vs virat kholi carrer comparison')
plt.xlabel('Season')
plt.ylabel('Runs scored')
```

```python

plt.plot(batsman['index'],batsman['V Kohli'],linestyle='dashdot',linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'],linestyle='dashdot',linewidth=1)
plt.title('Rohit sharma vs virat kholi carrer comparison')
plt.xlabel('Season')
plt.ylabel('Runs scored')
```

```python
batsman.head()
```

```python
#there are different markers and markersize
plt.plot(batsman['index'],batsman['V Kohli'],linestyle='dashdot',color='red',linewidth=3,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],linestyle='dashdot',color='blue',linewidth=1,marker='D')
plt.title('Rohit sharma vs virat kholi carrer comparison')
plt.xlabel('Season')
plt.ylabel('Runs scored')
```

```python
#Labeling indiviual line to know which line represents what
#there are different markers and markersize
plt.plot(batsman['index'],batsman['V Kohli'],linestyle='dashdot',color='red',linewidth=3,marker='D',markersize=10,label='Virat')#add lable
plt.plot(batsman['index'],batsman['RG Sharma'],linestyle='dashdot',color='blue',linewidth=1,marker='D',label='Rohit')  #add label
plt.title('Rohit sharma vs virat kholi carrer comparison')
plt.xlabel('Season')
plt.ylabel('Runs scored')

plt.legend()#call this function
```
