```python
df= pd.read_csv('gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'])
#tdoesnot show percentage
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')
plt.show()
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%' ,colors=['blue','red','yellow','pink','brown'])
plt.show()
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.1,0,0,0,0,0])
plt.show()
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.3,0,0,0,0,0])
plt.show()
```

```python

plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.1,0,0,0,0,0.3])
plt.show()
```

```python
plt.style.use('fivethirtyeight')#add the style at first
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.3,0,0,0,0,0])
plt.show()
```

```python
plt.style.use('ggplot')#add the style at first
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.3,0,0,0,0,0])
plt.show()
```
