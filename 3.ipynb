```python
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2,color='red')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='blue')
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2,color='yellow')
plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.show()
```

```python
#explanations
# Create first bar chart for 2015 data
# np.arange(df.shape[0]) creates positions 0,1,2,... based on number of rows in dataframe
# Subtract 0.2 to shift these bars slightly left
# Width=0.2 makes each bar take up 0.2 units of width
# Color is set to red
plt.bar(np.arange(df.shape[0])-0.2, df['2015'], width=0.2, color='red')

# Create second bar chart for 2016 data
# Uses same positions without shifting (centered)
# Same width, but colored blue
plt.bar(np.arange(df.shape[0]), df['2016'], width=0.2, color='blue')

# Create third bar chart for 2017 data
# Shifts bars 0.2 units to the right
# Same width, colored yellow
plt.bar(np.arange(df.shape[0])+0.2, df['2017'], width=0.2, color='yellow')

# Set the x-axis tick labels to show batsman names
# np.arange(df.shape[0]) creates the positions for the ticks
# df['batsman'] provides the labels to show at those positions
plt.xticks(np.arange(df.shape[0]), df['batsman'])

# Display the complete plot
plt.show()
```

```python
# Define a list of numbers representing the number of children
children = [10,20,40,10,30]

# Define a list of color names that will be used as x-axis labels
# Each color name is repeated multiple times in a string
colors = ['red red red red red red',
          'blue blue blue blue',
          'green green green green green',
          'yellow yellow yellow yellow ',
          'pink pinkpinkpink']

# Create a bar plot where:
# - x-axis shows the color names (colors list)
# - y-axis shows the number of children (children list)
# - All bars are colored black regardless of the color names
plt.bar(colors, children, color='black')

# Rotate the x-axis labels (the color names) vertically for better readability
plt.xticks(rotation='vertical')
```
